```python
from source.parser import Parser
import source.draw_molecule as draw
import matplotlib
import matplotlib.pyplot as plt
```

## Single Bonds

```python
molecule = Parser().parse_smiles('CCCNC')
graph = draw.draw_molecule(molecule)
```

## Multi Bonds

```python
molecule = Parser().parse_smiles('O$C=C#O')
graph = draw.draw_molecule(molecule)                            
```

## Aromatic Ring

```python
molecule = Parser().parse_smiles('c1ccccc1')
graph = draw.draw_molecule(molecule)  
```

## Branches

```python
molecule = Parser().parse_smiles('NCC(=CO)C')
graph = draw.draw_molecule(molecule)  
```

## Complicated Molecule
